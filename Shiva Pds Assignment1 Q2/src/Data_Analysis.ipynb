```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("../data_clean/clean_raw_data.csv")
df
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df['math score'], bins=20, kde=True)
plt.title('Histogram of scores')
plt.xlabel('math scores')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.scatterplot(x='math score', y='reading score', data=df)
plt.title('Scatter Plot of math score vs. reading score')
plt.xlabel('math score')
plt.ylabel('reading score')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='parental level of education', data=df)
plt.title('Bar Chart of parental level of education')
plt.xlabel('parental level of education')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.boxplot(x='gender', y='writing score', data=df)
plt.title('Boxplot of Writing score by Gender')
plt.xlabel('Gender')
plt.ylabel('Writing score')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.lineplot(x='race/ethnicity', y='reading score', data=df)
plt.title('Line Chart of  Scores Over race/ethnicity')
plt.xlabel('race/ethnicity')
plt.ylabel('Reading Scores')
plt.show()
```
